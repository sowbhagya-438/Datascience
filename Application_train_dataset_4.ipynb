```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Calculate Credit to Income Ratio
df['CREDIT_INCOME_RATIO'] = df['AMT_CREDIT'] / df['AMT_INCOME_TOTAL']

# Step 2: Group by education level and calculate average ratio
edu_credit_ratio = df.groupby('NAME_EDUCATION_TYPE')['CREDIT_INCOME_RATIO'].mean().reset_index()

# Step 3: Plot the bar graph
plt.figure(figsize=(10, 6))
sns.barplot(data=edu_credit_ratio, x='NAME_EDUCATION_TYPE', y='CREDIT_INCOME_RATIO', palette='viridis')
plt.title('Average Credit to Income Ratio by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Average Credit to Income Ratio')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create pivot table for average AMT_CREDIT
pivot_table = df.pivot_table(
    index='NAME_FAMILY_STATUS',
    columns='NAME_INCOME_TYPE',
    values='AMT_CREDIT',
    aggfunc='mean'
)

# Step 2: Plot the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap='YlGnBu')
plt.title('Average Loan Amount by Family Status and Income Type')
plt.xlabel('Income Type')
plt.ylabel('Family Status')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pivot_table = df.pivot_table(
    values='TARGET',
    index='NAME_FAMILY_STATUS',
    columns='NAME_INCOME_TYPE',
    aggfunc='mean'
)
pivot_table = pivot_table.round(3)
print(pivot_table)
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, fmt=".2%", cmap="Reds")
plt.title("Default Rate by Family Status and Income Type")
plt.ylabel("Family Status")
plt.xlabel("Income Type")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['AGE_YEARS'] = (-df['DAYS_BIRTH']) / 365
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=df,
    x='AGE_YEARS',
    y='AMT_CREDIT',
    hue='TARGET',
    palette={0: 'blue', 1: 'red'},
    alpha=0.5
)

plt.title("Scatter Plot: Age vs Loan Amount by Default Status")
plt.xlabel("Age (Years)")
plt.ylabel("Loan Amount (AMT_CREDIT)")
plt.legend(title="Default Status", labels=["No Default", "Default"])
plt.show()
```

```python
df_car_owners = df[df['OWN_CAR_AGE'].notna()]
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.histplot(data=df_car_owners, x='OWN_CAR_AGE', hue='TARGET', bins=20, kde=True, element="step")
plt.title("Distribution of Car Age by Default Status")
plt.xlabel("Car Age (Years)")
plt.ylabel("Count")
plt.legend(title="TARGET", labels=["No Default", "Default"])
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_filtered = df.dropna(subset=['EXT_SOURCE_2', 'NAME_CONTRACT_TYPE'])
g = sns.FacetGrid(df_filtered, col='NAME_CONTRACT_TYPE', height=5, aspect=1.2)
g.map(sns.histplot, 'EXT_SOURCE_2', kde=True, bins=30, color='skyblue')
g.fig.suptitle("Distribution of EXT_SOURCE_2 by Loan Type", y=1.05)
g.set_axis_labels("EXT_SOURCE_2", "Count")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_filtered = df.dropna(subset=['CNT_FAM_MEMBERS', 'AMT_ANNUITY'])

plt.figure(figsize=(10, 6))
sns.regplot(
    data=df_filtered,
    x='CNT_FAM_MEMBERS',
    y='AMT_ANNUITY',
    scatter_kws={'alpha': 0.4},
    line_kws={'color': 'red'}
)

plt.title("Loan Annuity vs. Family Size")
plt.xlabel("Number of Family Members")
plt.ylabel("Loan Annuity Amount (AMT_ANNUITY)")
plt.show()
```
