```python

df_filtered = df[(df['DAYS_EMPLOYED'] != 365243) & df['DAYS_EMPLOYED'].notna()]
df_filtered['YEARS_EMPLOYED'] = (-df_filtered['DAYS_EMPLOYED']) / 365
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.regplot(
    data=df_filtered,
    x='YEARS_EMPLOYED',
    y='AMT_CREDIT',
    scatter_kws={'alpha': 0.3},
    line_kws={'color': 'red'}
)

plt.title("Average Loan Amount vs. Years of Employment")
plt.xlabel("Years Employed")
plt.ylabel("Loan Amount (AMT_CREDIT)")
plt.show()

```

```python

df_filtered = df[(df['DAYS_EMPLOYED'] != 365243) & df['EXT_SOURCE_3'].notna()]
df_filtered['YEARS_EMPLOYED'] = (-df_filtered['DAYS_EMPLOYED']) / 365
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.regplot(
    data=df_filtered,
    x='YEARS_EMPLOYED',
    y='EXT_SOURCE_3',
    scatter_kws={'alpha': 0.3},
    line_kws={'color': 'green'}
)

plt.title("EXT_SOURCE_3 vs. Employment Duration")
plt.xlabel("Years Employed")
plt.ylabel("EXT_SOURCE_3 Score")
plt.show()

```

```python

NAME_INCOME_TYPE_defaults = df.groupby('NAME_INCOME_TYPE')['TARGET'].agg(['count', 'sum'])
NAME_INCOME_TYPE_defaults.rename(columns={'count': 'Total_Applications', 'sum': 'Defaults'}, inplace=True)
NAME_INCOME_TYPE_defaults['Default_Rate'] = NAME_INCOME_TYPE_defaults['Defaults'] / NAME_INCOME_TYPE_defaults['Total_Applications']
NAME_INCOME_TYPE_defaults_sorted = NAME_INCOME_TYPE_defaults.sort_values(by='Default_Rate', ascending=False)
print(NAME_INCOME_TYPE_defaults_sorted.head(10))
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.barplot(data=NAME_INCOME_TYPE_defaults_sorted.reset_index().head(10),
            x='NAME_INCOME_TYPE', y='Default_Rate', palette='Reds_r')
plt.xticks(rotation=45)
plt.title("Top 10 NAME_INCOME_TYPES with Highest Default Rates")
plt.ylabel("Default Rate")
plt.xlabel("NAME_INCOME_TYPE")
plt.tight_layout()
plt.show()
  # Top 10 highest default rate regions
```

```python
import pandas as pd

default_by_income = df.groupby('NAME_INCOME_TYPE')['TARGET'].mean().sort_values(ascending=False)

counts = df['NAME_INCOME_TYPE'].value_counts()
summary = pd.DataFrame({
    'Default Rate': default_by_income,
    'Applicant Count': counts
})

print(summary)
import seaborn as sns
import matplotlib.pyplot as plt

# Bar plot
plt.figure(figsize=(10,6))
sns.barplot(y=default_by_income.values, x=default_by_income.index, palette='viridis')
plt.title('Default Rate by Income Type')
plt.xlabel('Income Type')
plt.ylabel('Default Rate')
plt.grid(axis='x')
plt.show()

```

```python

df['ANNUITY_INCOME_RATIO'] = df['AMT_ANNUITY'] / df['AMT_INCOME_TOTAL']

housing_annuity_ratio = df.groupby('NAME_HOUSING_TYPE')['ANNUITY_INCOME_RATIO'].mean().reset_index()
housing_annuity_ratio = housing_annuity_ratio.sort_values(by='ANNUITY_INCOME_RATIO', ascending=False)
print(housing_annuity_ratio)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(data=housing_annuity_ratio,
            x='NAME_HOUSING_TYPE', y='ANNUITY_INCOME_RATIO', palette='Purples_d')
plt.title('Average Annuity to Income Ratio by Housing Type')
plt.xlabel('Housing Type')
plt.ylabel('Average Annuity / Income Ratio')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

df['ANNUITY_CREDIT_RATIO'] = df['AMT_ANNUITY'] / df['AMT_CREDIT']
df_filtered = df[df['ANNUITY_CREDIT_RATIO'] < 0.1]
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.histplot(data=df_filtered, x='ANNUITY_CREDIT_RATIO', bins=100, color='teal')
plt.title('Distribution of Annuity to Credit Ratio')
plt.xlabel('Annuity / Credit')
plt.ylabel('Number of Applicants')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df_filtered = df[df['EXT_SOURCE_1'].notna()]


df_filtered['EXT_SOURCE_1_BIN'] = pd.qcut(df_filtered['EXT_SOURCE_1'], q=10, duplicates='drop')


grouped = df_filtered.groupby(['NAME_CONTRACT_TYPE', 'EXT_SOURCE_1_BIN'])['TARGET'].mean().reset_index()
grouped.rename(columns={'TARGET': 'Default_Rate'}, inplace=True)
plt.figure(figsize=(12, 5))
sns.lmplot(data=df_filtered, x='EXT_SOURCE_1', y='TARGET', hue='NAME_CONTRACT_TYPE', lowess=True)
plt.title('Default Rate vs. EXT_SOURCE_1 Score by Loan Type')
plt.xlabel('EXT_SOURCE_1 (Binned)')
plt.ylabel('Default Rate')
plt.xticks(rotation=45)
plt.legend(title='Loan Type')
plt.tight_layout()
plt.show()

```

```python
# Group by organization type and compute default rate
org_risk = df.groupby('ORGANIZATION_TYPE')['TARGET'].mean().reset_index()
org_risk = org_risk.sort_values(by='TARGET', ascending=False)

# Get the organization type with the highest default rate
highest_risk_org = org_risk.iloc[0]['ORGANIZATION_TYPE']
print(f"Highest Risk Organization Type: {highest_risk_org}")
# Filter applicants from the riskiest organization type
df_risk_group = df[df['ORGANIZATION_TYPE'] == highest_risk_org].copy()

# Convert DAYS_BIRTH to age in years
df_risk_group['AGE_YEARS'] = -df_risk_group['DAYS_BIRTH'] / 365
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.histplot(df_risk_group['AGE_YEARS'], bins=30, kde=True, color='coral')
plt.title(f'Age Distribution of Applicants from {highest_risk_org}')
plt.xlabel('Age (Years)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df_risk_group['AMT_INCOME_TOTAL'], bins=40, kde=True, color='seagreen')
plt.title(f'Income Distribution of Applicants from {highest_risk_org}')
plt.xlabel('Annual Income')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
