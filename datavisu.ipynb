```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.DataFrame(np.random.randint(0,100,10),columns=['sales'])
```

```python
df.head()
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,7))
```

```python
plt.plot(df['sales'],marker='o',color='blue')
```

```python
plt.plot(df['sales'],marker='o',color='blue',markerfacecolor='pink')
```

```python
plt.plot(df['sales'],marker='o',color='blue',markerfacecolor='pink',markeredgecolor='green')
```

```python
plt.plot(df['sales'],marker='o',color='blue',markerfacecolor='pink',markeredgecolor='green',markeredgewidth=5)
```

```python
plt.plot(df['sales'],marker='o',color='blue',markerfacecolor='pink',markeredgecolor='green',markeredgewidth=5,linewidth=10)
```

```python
plt.plot(df['sales'],marker='o',color='blue',markerfacecolor='pink',markeredgecolor='green',markeredgewidth=5,linewidth=10,alpha=0.9)
```

```python
plt.xlabel("Index")
```

```python
plt.ylabel("sales")
```

```python
plt.title("sales vs Index")
```
